```python
import pandas as pd
import pymongo
import json
```

```python
client = pymongo.MongoClient("mongo")
database = client['exercices']
collection = database['youtube']
```

```python
df_youtube = pd.read_csv("./data/USvideos.csv")
df_youtube.head(2)
```

## Question 0

```python
dic1 = df_youtube.to_dict("records")
dic1
```

```python
with open("./data/US_category_id.json", "r") as read_file:
    dic2 = json.load(read_file)
```

```python
dic2
```

### Netoyer les données

```python
dic1[1]['tags']#.split("|")
```

```python
#pour séparer en listes les tags
for i in dic1:
    i['tags'] = i['tags'].split("|")
```

```python
#pour rajouter le nom des catégories
for i in dic1:
    for j in dic2['items']:
        if i['category_id'] == int(j['id']) :
            i['category_name'] = j['snippet']['title']
    
    
```

```python
collection.delete_many({})
```

## Question 5

```python
cur = collection.aggregate([{"$group" : {"_id" : "$category_name", "AverageViewsByCategory" : {"$avg" :"$views"}}}])
list(cur)
```

## Question 6 

```python
cur = collection.aggregate([
    {"$group" : {"_id" : "$channel_title", "LikeByChannel" : {"$avg" :"$likes"}}},
    { '$sort': { "LikeByChannel": -1 }},
    { '$limit' : 10 }
    ])
list(cur)
```
